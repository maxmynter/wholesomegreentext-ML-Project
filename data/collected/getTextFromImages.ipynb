```python
import cv2
import pytesseract
import os
import json
```

```python
def get_text_from_image(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding for high-contrast images
    _, thresh_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)

    # Extract text using Tesseract OCR
    extracted_text = pytesseract.image_to_string(thresh_image)

    return extracted_text


def process_images_in_directory(directory_path):
    text_data = {}  # Dictionary to store text data with image filenames as keys
    n_max = len(os.listdir(directory_path))
    for i, filename in enumerate(os.listdir(directory_path)) :
        print(i/ n_max)
        if filename.endswith(('.jpg', '.png', '.jpeg')):
            image_path = os.path.join(directory_path, filename)

            # Read the image using OpenCV
            image = cv2.imread(image_path)

            # Get text from the image
            extracted_text = get_text_from_image(image)

            # Store the extracted text in the dictionary
            text_data[filename] = extracted_text

    return text_data
```

```python
output_file = '../data/raw_extracted_text.json'
image_directory = '../data/images/'
text_data = process_images_in_directory(image_directory)
with open(output_file, 'w') as json_file:
    json.dump(text_data, json_file)
```
